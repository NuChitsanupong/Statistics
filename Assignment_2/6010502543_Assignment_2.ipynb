```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
RMSE = sqrt(mean_squared_error(df2.T_1200_1430, df2.predict))
RMSE
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
MAPE = mean_absolute_percentage_error(df2.T_1200_1430, df2.predict)
MAPE
```

```python
df_pax = pd.read_csv("Regression_Data-pax.csv")
df_pax
```

## min-max normalization

```python
df_paxx = df_pax.iloc[:,5:].copy()
df_paxx.head()
```

```python
normalized_df_paxx=(df_paxx-df_paxx.min())/(df_paxx.max()-df_paxx.min())
normalized_df_paxx.head()
```

```python
df_pax = pd.concat([df_pax.iloc[:,:5],normalized_df_paxx], axis=1)
```

```python
df_pax
```

```python
df_pax_pred = df_pax.copy()
```

```python
X = df_pax.iloc[:,5:]
X = sm.add_constant(X, prepend=False)
Y = df_pax['#waiting passengers (y)']
model = sm.OLS(Y, X)
resultss = model.fit()
df_pax_pred = df_pax_pred.assign(predict_waiting_passengers = resultss.predict(X))
print(resultss.summary())
```

```python
X = df_pax.iloc[:,5:]
X = sm.add_constant(X, prepend=False)
Y = df_pax['#departing passengers (y)']
model = sm.OLS(Y, X)
results = model.fit()
df_pax_pred = df_pax_pred.assign(predict_departing_passengers = results.predict(X))
print(results.summary())
```

```python
X = df_pax.iloc[:,5:]
X = sm.add_constant(X, prepend=False)
Y = df_pax['#arriving passengers (y)']
model = sm.OLS(Y, X)
results = model.fit()
df_pax_pred = df_pax_pred.assign(predict_arriving_passengers = results.predict(X))
print(results.summary())
```

```python
df_pax_pred
```

```python
RMSE_waiting_passengers = sqrt(mean_squared_error(df_pax_pred['#waiting passengers (y)'], df_pax_pred['predict_waiting_passengers']))
RMSE_waiting_passengers
```

```python
RMSE_departing_passengers = sqrt(mean_squared_error(df_pax_pred['#departing passengers (y)'], df_pax_pred['predict_departing_passengers']))
RMSE_departing_passengers
```

```python
RMSE_arriving_passengers = sqrt(mean_squared_error(df_pax_pred['#arriving passengers (y)'], df_pax_pred['predict_arriving_passengers']))
RMSE_arriving_passengers
```

## Hypothesis test

```python
df_thailand_goods = df_thailand[['Year', 'goodsImports', 'goodsExports']]
df_thailand_goods.head()
```

```python
df_thailand_goods.set_index('Year', inplace=True)
df_thailand_goods.head()
```

```python
df_thailand_goods.describe()
```

```python
meanx = df_thailand_goods["goodsImports"].mean()
meany = df_thailand_goods["goodsExports"].mean()
stdx = df_thailand_goods["goodsImports"].std()
stdy = df_thailand_goods["goodsExports"].std()
countx = df_thailand_goods["goodsImports"].count()
county = df_thailand_goods["goodsExports"].count()

print("meanx = " + str(meanx))
print("meany = " + str(meany))
print("stdx = " + str(stdx))
print("stdy = " + str(stdy))
print("countx = " + str(countx))
print("county = " + str(county))
```

```python
mean = meanx-meany
mean
```

```python
sd = np.sqrt((pow(stdx,2)/countx)+(pow(stdy,2)/county))
sd
```

```python
test_statistics = (mean-0)/sd
test_statistics
```

```python
x = (pow(stdx,2)/countx)
y = (pow(stdy,2)/county)
```

```python
import math
degree_of_freedom = math.floor(pow(x+y,2)/((pow(x,2)/countx-1)+(pow(y,2)/county-1)))
degree_of_freedom
```

```python
from scipy.stats import t
alpha = 0.05/2
t.ppf(1-alpha, degree_of_freedom)
```

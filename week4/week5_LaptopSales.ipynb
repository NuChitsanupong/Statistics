```python
newmeanx = df_100[df_100.HDSize == 40]["RetailPrice"].mean()
newmeany = df_100[df_100.HDSize == 80]["RetailPrice"].mean()
newstdx = df_100[df_100.HDSize == 40]["RetailPrice"].std()
newstdy = df_100[df_100.HDSize == 80]["RetailPrice"].std()
newcountx = df_100[df_100.HDSize == 40]["RetailPrice"].count()
newcounty = df_100[df_100.HDSize == 80]["RetailPrice"].count()
```

```python
newmean = newmeanx-newmeany
newmean
```

```python
newsd = np.sqrt((pow(newstdx,2)/newcountx)+(pow(newstdy,2)/newcounty))
newsd
```

```python
new_test_statistics = (newmean-0)/newsd
new_test_statistics
```

```python
x = (pow(newstdx,2)/newcountx)
y = (pow(newstdy,2)/newcounty)
```

```python
degree_of_freedom = math.floor(pow(x+y,2)/((pow(x,2)/newcountx-1)+(pow(y,2)/newcounty-1)))
degree_of_freedom
```

```python
alpha = 0.05/2
t.ppf(1-alpha, degree_of_freedom)
```

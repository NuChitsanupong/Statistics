```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as pyo
import plotly.graph_objs as go
import re
%matplotlib inline
```

```python
df = pd.read_csv('railLine_UrbanPopulation.csv',sep=',',header='infer')
df
```

```python
IQR = df.quantile(0.75) - df.quantile(0.25)
IQR

df = df[(df.railLine2017 <= df.railLine2017.quantile(0.75) + 1.5*IQR['railLine2017']) 
    & (df.railLine2017 >= df.railLine2017.quantile(0.25) - 1.5*IQR['railLine2017'])]
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),
            square=True,
            linewidths=0.25,    
            linecolor=(0,0,0),
           cmap=sns.color_palette("coolwarm"),
           annot=True)
```

# Regions

```python
df.Region.unique()
```

```python
region_name = {'W_Eu':'Western Europe',
               'W_As':'Western Asia',
               'E_Eu':'Eastern Europe',
               'E_As':'East Asia',
               'Afr':'Africa',
               'N_Eu':'Northern Europe',
               'M_Es':'Middle East',
               'SE_As':'Southeast Asia'}
```

```python
dic = {'W_Eu' : df[df.Region == 'Western Europe'],
'W_As' : df[df.Region == 'Western Asia'],
'E_Eu' : df[df.Region == 'Eastern Europe'],
'E_As' : df[df.Region == 'East Asia'],
'Afr'  : df[df.Region == 'Africa'],
'N_Eu' : df[df.Region == 'Northern Europe'],
'M_Es' : df[df.Region == 'Middle East'],
'SE_As' : df[df.Region == 'Southeast Asia']}
```

```python
railLine = {}
urban_population = {}
x = 0
for key in dic:
    y = []
    z = []
    for i in dic[key].sum():
        if x>2 and x<=11:
            y.append(i)
        elif x>11:
            z.append(i)
        x += 1
    railLine[key] = y
    urban_population[key] = z
    x = 0
```

```python
#railLine
#urban_population
```

```python
year = ['2000','2010','2011','2012','2013','2014','2015','2016','2017']
```

```python
region_data = {}
for key in dic:
    data = {'railLine' : railLine[key],
            'urban_population' : urban_population[key],
            'year' : year}
    df1 = pd.DataFrame(data)
    df1.set_index('year', inplace=True)
    region_data[key] = df1
```

# Time-series

# RailLine

```python
data = []
for key in region_data:
    trace = go.Scatter(x=region_data[key].index,
                        y=region_data[key]['railLine'],
                        mode='lines+markers',
                        name=region_name[key])
    data.append(trace)
```

```python
layout = go.Layout(title='2000-2017 Rail Line',
                   xaxis = {'title':'Year'},
                   yaxis = {'title':'Distance[Km]',
                           'range': [0,100000]}, 
                   hovermode='closest',
                   )
```

```python
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

# Urban_population

```python
data = []
for key in region_data:
    trace = go.Scatter(x=region_data[key].index,
                        y=region_data[key]['urban_population'],
                        mode='lines+markers',
                        name=region_name[key])
    data.append(trace)
```

```python
layout = go.Layout(title='2000-2017 Urban population',
                   xaxis = {'title':'Year'},
                   yaxis = {'title':'population',
                           'range': [10000000,170000000]}, 
                   hovermode='closest',
                   )
```

```python
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
